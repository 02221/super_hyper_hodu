### 블랙 숄즈 머튼 모형을 따르는 동적인 주가 시뮬레이션  

$S_t = S_{t-dt} exp((r- \frac{1}{2} \sigma^2) dt + \sigma \sqrt dt z_t)$

```python
S0 = 100 # 초기 주가
r = 0.05 # 무위험 단기 이자율
sigma = 0.25 # 고정 변동성
T = 2.0 # 시간 (년)
I = 1000 # 난수의 수
M = 50 # 만기
dt = T / M
S = np.zeros((M + 1, I)) #주가
S[0] = S0

for t in range(1, M + 1):
    S[t] = S[t-1] * np.exp((r - 0.5 * sigma ** 2) * dt
    + sigma * np.sqrt(dt) * npr.standard_normal(I))
```

```python
plt.hist(S[-1], bins=50)
plt.xlabel('index level')
plt.ylabel('frequency')
plt.title('GBM value at the simulation maturity', fontsize=20)
plt.grid(True)
```

```python
plt.plot(S[:, :10], lw = 1.5)
plt.xlabel('time')
plt.ylabel('index level')
plt.title('GBM simulation path (only 10 shown)', fontsize=20)
plt.grid(True)
```
